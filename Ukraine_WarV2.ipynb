```python
# Definir o tema branco
sns.set_theme(style='whitegrid')

# Calcular as médias das colunas
medias = ukraine_war_cleaned_df.mean()

# Remover as colunas especificadas
colunas_a_remover = ['day', 'APC', 'MRL', 'personnel', 'ArtilhariaAeronautica', 'ArtilhariaTerrestre', 'ArtilhariaNaval']
medias_filtradas = medias.drop(colunas_a_remover)

# Renomear as colunas para o gráfico ficar mais intuitivo
traducao_colunas = {
    'aircraft': 'Aeronave',
    'helicopter': 'Helicoptero',
    'tank': 'Tanque',
    'field artillery': 'ArtilhariaTerrestre',
    'drone': 'Drone',
    'naval ship': 'NavioNaval',
    'anti-aircraft warfare': 'GuerraAntiaerea',
    'special equipment': 'EquipamentoEspecial',
    'submarines': 'Submarinos'
}
medias_filtradas = medias_filtradas.rename(traducao_colunas)

# Criar um gráfico de teia
categorias = medias_filtradas.index
valores = medias_filtradas.values

categorias = np.append(categorias, categorias[0])
valores = np.append(valores, valores[0])
angulos = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False)

# Criar o gráfico de teia
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plotar os dados
ax.plot(angulos, valores, marker='o', color='skyblue', label='Médias')
ax.fill(angulos, valores, color='skyblue', alpha=0.25)

# Adicionar rótulos para cada categoria sem espaços
ax.set_xticks(angulos)
ax.set_xticklabels(categorias, color='grey', size=10, rotation=45, ha='right')

# Mover as legendas para fora do gráfico
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

plt.title('Gráfico de Teia das Médias das Colunas do DataFrame', size=15)
plt.show()
```

```python
# Criar uma cópia do DataFrame com as colunas relevantes
df_artilharia = ukraine_war_cleaned_df[['date', 'ArtilhariaAeronautica', 'ArtilhariaTerrestre', 'ArtilhariaNaval']].copy()

# Agrupar por data e somar o uso de artilharia em cada categoria
df_artilharia = df_artilharia.groupby('date').sum().reset_index()

# Definir o tema branco
sns.set_theme(style='whitegrid')

# Criar um gráfico de linha
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='ArtilhariaAeronautica', data=df_artilharia, label='Artilharia Aeronáutica')
sns.lineplot(x='date', y='ArtilhariaTerrestre', data=df_artilharia, label='Artilharia Terrestre')
sns.lineplot(x='date', y='ArtilhariaNaval', data=df_artilharia, label='Artilharia Naval')

# Configurações do gráfico
plt.title('Uso Total da Artilharia ao Longo do Tempo', size=15)
plt.xlabel('Data')
plt.ylabel('Quantidade Total')
plt.legend()
plt.show()
```

```python
# Dados fornecidos
dados_equipamentos = {'Aeronave': 329.0,
                      'Helicoptero': 324.0,
                      'Tanque': 5977.0,
                      'ArtilhariaTerrestre': 8464.0,
                      'MRL': 943.0,
                      'Drone': 6591.0,
                      'NavioNaval': 23.0,
                      'GuerraAntiaerea': 623.0,
                      'EquipamentoEspecial': 1268.0,
                      'Submarinos': 1.0}

# Criando um DataFrame a partir dos dados
df_equipamentos = pd.DataFrame(list(dados_equipamentos.items()), columns=['Equipamento', 'Quantidade'])
df_equipamentos = df_equipamentos.sort_values(by='Quantidade')

# Definir o tema branco
sns.set_theme(style='whitegrid')

# Criando o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando as barras
ax.bar(df_equipamentos['Equipamento'], df_equipamentos['Quantidade'], color='skyblue')

# Rótulos e título
ax.set_xlabel('Equipamento')
ax.set_ylabel('Quantidade')
ax.set_title('Distribuição de Equipamentos de Guerra')

# Rotacionando os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Adicionando os valores nas barras
for i, v in enumerate(df_equipamentos['Quantidade']):
    ax.text(i, v + 50, str(int(v)), color='black', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
